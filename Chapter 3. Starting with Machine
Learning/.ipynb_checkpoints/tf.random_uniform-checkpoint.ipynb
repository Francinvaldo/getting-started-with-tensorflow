```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
numvem_de_pontos = 500
#y = 0.22x + 0.78:
x_p = []
y_p = []
a = 0.22
b = 0.78

for i in range(numvem_de_pontos):
    x = np.random.normal(0.0 , 1.0 )
    y =  a*x + b + np.random.normal(0.0 , 0.1)
    x_p.append([x])
    y_p.append([y])  
    
plt.plot(x_p , y_p  ,'p' , label = 'input data')
plt.legend()
plt.show()
```
